5. Find the top 5 customers who spent the most on purchases.

```python
cursor = conn.execute('''
with txn_table as (
  select
  cust_id,
  sum(txn_value) as txn_1
  from txns group by 1
)

select * from (select
  *,
  rank() over (order by txn_1 desc) as rnk
from txn_table )
where rnk <= 5
''')

for row in cursor:
  print(row)
```

6. Find the customers who have made purchases in consecutive days

```python
cursor = conn.execute('''
with
cte1 as (
select
  cust_id,
  txn_dt,
  cast(julianday(txn_dt)-julianday('1990-01-01') as int) - row_number() over (order by txn_dt asc) as date_rn,
  sum(txn_value)

from txns
where cust_id = 'saravana'
group by 1,2)

select
  date_rn,
  min(txn_dt),
  max(txn_dt),
  count(*)
from cte1
group by 1
''')


for row in cursor:
  print(row)
```

8. List the customers who have made purchases in both the current and previous year

```python
cursor = conn.execute('''
with
cy as (
select *
from txns
where cast(strftime('%y',txn_dt) as int) = 2022
),
py as (
select *
from txns
where cast(strftime('%y',txn_dt) as int) = 2021
)

select
cy.cust_id
from cy join py
on cy.cust_id = py.cust_id
group by 1
''')


for row in cursor:
  print(row)
```

10. List the customers who have made purchases in every month of the year

18. Write an sql query to find the Average time taken for a customer to Repeat a Purchase

```python
cursor = conn.execute('''
with cte1 as (
select
  cust_id,
  txn_dt,
  julianday(txn_dt) ju_dt,
  row_number() over (partition by cust_id order by txn_dt asc) as rn
from txns
where cust_id = 'saravana'
),
cte2 as (
select
  *,
  (select (t2.ju_dt - t1.ju_dt) from cte1 t2
where t2.rn - 1 = t1.rn) as flag
from cte1 t1 )

--select * from cte2
select cust_id,cast(sum(flag) as decimal(18,9))/cast((count(*)-1) as decimal(18,9)) from cte2 group by 1
''')


for row in cursor:
  print(row)
```

19. Find Employees who are earning more salaries than their Managers

```python
cursor = conn.execute('''
with manager as (
select
  b.name,
  b.salary as e_salary,
  b.manager,
  a.salary as m_salary
from employees a
join employees b
on a.name = b.manager
group by 1,2,3,4
)

select
*,
  case when e_salary > m_salary then 1 else 0 end as flag
from manager
''')
for row in cursor:
  print(row)
```
