Libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

Pie chart - matplotlib

```python
fruits = ['apple', 'mango', 'banana', 'orange']
supply = [20,30,40,10]

fig, ax = plt.subplots(2,3, figsize=(10,7))
ax[0,0].pie(supply, labels=fruits, autopct='%1.1f%%')
ax[0,1].pie(supply, labels=fruits, autopct='%1.1f%%', labeldistance=0.6, pctdistance=1.2)
ax[0,2].pie(supply, labels=fruits, autopct='%1.1f%%', labeldistance=0.6, pctdistance=1.2, explode=(0,0.2,0,0))
ax[1,0].pie(supply, labels=fruits, autopct='%1.1f%%', explode=(0,0.2,0,0), startangle=90, shadow=True)
ax[1,1].pie(supply, labels=fruits, autopct='%1.1f%%', wedgeprops={'width':0.4})
ax[1,2].pie(supply, labels=fruits, autopct='%1.1f%%', radius=0.5, textprops={'size':'smaller'})
fig.suptitle("Pie chart & different functionalities")
plt.show()

# nested pie chart
# bar of pie chart
```

Barplot using Matplotlib ( Vertical and Horizontal plot)

```python
# Bar plot - matplotlib

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
counts_2 = [20,20,20,20]
x=np.arange(4)
bar_colors = ['red', 'c', 'k','orange']


fig, ax = plt.subplots(2,2,figsize=(7,7))

# bar plot
ax[0,0].barh(fruits, counts, label=fruits, color=bar_colors)
ax[0,0].set(xlabel="",ylabel="",title="")
#ax[0,0].legend(title='Fruit color', loc='upper right', prop={'size':8})


# bar plot horizontal
ax[0,1].bar(fruits, counts, label=fruits, color=bar_colors)
ax[0,1].set(xlabel="",ylabel="",title="")
#ax[0,1].legend(title='Fruit color', loc='upper right', prop={'size':8})

# grouped bar plot
ax[1,0].bar(x-0.2, counts, label=fruits, color=bar_colors, width=0.35)
ax[1,0].bar(x+0.2, counts_2, label='', color=bar_colors, width=0.35)
ax[1,0].set(xlabel="",ylabel="",title="",xticks=x,xticklabels=fruits)
#ax[1,0].legend(title='Fruit color', loc='upper right', prop={'size':8})

# stacked bar plot
ax[1,1].bar(fruits, counts, bottom=counts_2, label='Top')
ax[1,1].bar(fruits, counts_2,label='bottom')
ax[1,1].set(xlabel="",ylabel="",title="")
ax[1,1].legend(title='label', loc='upper right', prop={'size':8})

#fig.legend(title='Fruit color', loc='upper right', prop={'size':8})
fig.suptitle("Fruit supplies")
plt.show()
```

Box plot - Matplotlib

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
rows=2
cols=3
colors=['red','blue','brown']

fig,ax = plt.subplots(rows,cols,figsize=(10,7))
ax[0,0].boxplot(data,patch_artist=True, vert=False)
ax[0,1].boxplot(data,patch_artist=True, showfliers=False)
ax[0,2].boxplot(data,patch_artist=True, notch=True, showfliers=True)

ax[1,0].boxplot(data,patch_artist=True)
ax[1,1].boxplot(data,patch_artist=True, boxprops={'facecolor':'r'}) #facecolor will work only with patch_artist=True, else error
ax[1,2].boxplot(data,patch_artist=True, boxprops={'facecolor':'brown'})

for i in range(rows): 
    for j in range(cols): 
        ax[i,j].xaxis.grid(True) if(i==0 and j==1) else ax[i,j].yaxis.grid(True)
fig.suptitle("Box plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime

dic = {
    'date': np.array([datetime.datetime(2023,1,i+1) for i in range(7)]),
    'value': np.random.randint(10, size=7)
}

df = pd.DataFrame(dic)
x = df['date']
y = df['value']
y_tick = df['value'] - 5



fig, ax = plt.subplots(2,2, figsize=(10,6))
ax[0,0].plot(x,y)
ax[0,0].plot(x,y_tick)
ax[0,1].plot(x,y, color='lightpink')
ax[1,0].plot(x,y, color='lightpink')
ax[1,1].plot(x,y, color='lightpink')
plt.show()
```

Scatter plot - matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

x = np.random.randint(100, size=100)
y = np.random.randint(100,size=100)
s = np.random.randint(200,size=100)

fig, ax = plt.subplots(2,2, figsize=(10,6))
ax[0,0].scatter(x,y)
ax[0,1].scatter(x,y, color='lightpink')
ax[1,0].scatter(x,y, color='lightpink', alpha=0.1)
ax[1,1].scatter(x,y, color='lightpink', alpha=0.5, s=s)
plt.show()
```
